```python
from llm.openai import OpenAIChat
```

```python
import os
from dotenv import load_dotenv
load_dotenv()
 default_llm = OpenAIChat(
                openai_api_key=os.getenv('OPENAI_API_KEY'),
                model_name="gpt-3.5-turbo-0125",
                max_tokens=4000
            )
```

```python
from unstructured_data_extractor import DataExtractor
```

```python
extractor = DataExtractor(llm=default_llm)
```

```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("RAG_document.pdf")
pages = loader.load_and_split()
print(pages[1].page_content)
```

```python
text = ""
for page in pages[10:20]:
    text += page.page_content
```

```python
import pickle

```

```python
import tiktoken
encoding = tiktoken.encoding_for_model("gpt-3.5-turbo-0125")
print(len(encoding.encode(text)))
```

```python
result = extractor.run(text)
```

```python
print(result["relationships"])
```

```python
print(result["nodes"])
```

```python
with open('extracted_result.pkl', 'wb') as file:
    # Serialize the dictionary and save it to the file
    pickle.dump(result, file)
```

```python
labels = set()
for node in result["relationships"]:
    labels.add(node["type"])
```
